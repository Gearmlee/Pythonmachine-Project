```python
def sort_months(months_str):
    months_list = list(map(int, months_str.split(', ')))
    months_list.sort()  # 정렬
    return ', '.join(map(str, months_list))  #변환
```

```python
def generate_info2(row):
    기본정보 = (f"{row['위치주소']}에 있는 체육시설의 ")

    if not pd.isnull(row['주소']) and not pd.isnull(row['시설이름']):
        기본정보 += f"상세 주소는 {row['주소']} {row['시설이름']}으로, "
    else:
        기본정보 += f"상세 주소는 {row['주소']}이며, "

    if not pd.isnull(row['우편번호']):
        기본정보 += (f"우편번호가 {row['우편번호']}이고, ")

    if not pd.isnull(row['종목명']):
        기본정보 += f"이용 가능한 종목은 {row['종목명']}입니다. "


    if pd.isnull(row['상세종목']):
        강좌정보 = "강좌가 존재하지 않습니다. "
    else:
        강좌정보 = (f"제공하는 강좌의 강좌명은 {row['상세종목']}이며, 강좌는 {row['개최월']}월에 개최되며, "
                     f"가격은 {row['가격']}원입니다. ")



    return (기본정보 +
            강좌정보)
```

```python
df_main = df_main.drop(columns=['정보'])
```

```python
qa_dataset = []

# 각 위치주소에 대해 반복합니다.
for 위치주소 in df_main['위치주소'].unique():
    # 각 위치주소에 대해 모든 종목명에 대해 반복합니다.
    for 종목 in df_main['종목명'].unique():
        # 해당 위치주소와 종목에 맞는 데이터를 필터링합니다.
        filtered_data = df_main[(df_main['위치주소'] == 위치주소) & (df_main['종목명'] == 종목)]

        # 필터링된 데이터가 있을 경우 Q&A를 생성합니다.
        if not filtered_data.empty:
            개수 = len(filtered_data)  # 강좌 개수
            질문 = f"{위치주소}에서 {종목}에 대한 강좌를 알려줘."

            if 개수 == 1:
                시설이름 = filtered_data.iloc[0]['시설이름']
                상세종목 = filtered_data.iloc[0]['상세종목']
                답변 = f"{위치주소}에는 \"{시설이름}\"에서 하는 \"{상세종목}\" 프로그램이 있습니다."
            else:
                답변 = f"{위치주소}에는 {종목} 강좌가 {개수}개 있습니다. 추가적으로 더 자세한 정보가 필요하시다면, \"{위치주소} {종목} 추가정보\" 텍스트를 입력해주세요."

            qa_dataset.append({'질문': 질문, '답변': 답변})
        else:
            # 종목에 대한 프로그램이 없을 경우
            질문 = f"{위치주소}에서 {종목}에 대한 강좌를 알려줘."
            답변 = f"{위치주소}에는 {종목}에 관한 프로그램이 존재하지 않습니다."
            qa_dataset.append({'질문': 질문, '답변': 답변})

# 결과를 데이터프레임으로 변환
qa_df = pd.DataFrame(qa_dataset)
```

```python
qa_dataset2 = []  # 추가 정보 요청을 위한 데이터셋

# 각 위치주소에 대해 반복합니다.
for 위치주소 in df_main['위치주소'].unique():
    # 각 위치주소에 대해 모든 종목명에 대해 반복합니다.
    for 종목 in df_main['종목명'].unique():
        # 해당 위치주소와 종목에 맞는 데이터를 필터링합니다.
        filtered_data = df_main[(df_main['위치주소'] == 위치주소) & (df_main['종목명'] == 종목)]

        # 필터링된 데이터가 있을 경우 추가 정보 요청을 위한 Q&A를 생성합니다.
        if not filtered_data.empty:
            개수 = len(filtered_data)  # 강좌 개수

            # 강좌가 2개 이상인 경우에만 추가 정보 요청을 생성
            if 개수 > 1:
                시설_상세정보 = []
                for index, row in filtered_data.iterrows():
                    시설이름 = row['시설이름']
                    상세종목 = row['상세종목']
                    # 큰따옴표를 사용하여 문자열 생성
                    시설_상세정보.append(f"\"{시설이름}\"에서 하는 \"{상세종목}\" 프로그램이 있습니다.")

                # 추가 정보 요청 질문 및 답변 생성
                질문 = f"{위치주소} {종목} 추가정보"
                답변 = "\n".join(시설_상세정보)
                qa_dataset2.append({'질문': 질문, '답변': 답변})

# 추가 정보 요청 데이터프레임 생성
qa_df2 = pd.DataFrame(qa_dataset2)

# 추가 정보 데이터셋 확인
print(qa_df2)
```

```python
lastadata_double = pd.concat([qa_df, qa_df2], ignore_index=True)

#lastadata_double.to_csv('/content/drive/MyDrive/lastadata_double.csv', index=False, encoding='euc-kr')
```
